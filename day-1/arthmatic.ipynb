```python
#swap two numbers
a=10
b=20
temp=a
a=b
b=temp
print(a,b)

```

```python
#swap two numbers without third variable
a=10
b=20
a,b = b,a
print(a,b)
```

```python
#square of number
a=10
print(a*a)
print(a**2)
```

```python
#cube of number
a=5
print(a*a*a)
print(a**3)
```

```python
#quotient and  remainder 
a = 16
b = 3

print("Quotient =", a // b)
print("Remainder =", a % b)
```

```python
#average of two numbers
a = 15
b = 25

avg = int((a + b) / 2)

print(avg)
```

```python
#simple interest 
p = 100
r = 2
t = 30

si = (p * r * t) / 100

print(si)
```

```python
#compound interest
p = 100
r = 2
t = 30

amount = p * (1 + r / 100) ** t

print(amount - p)
```

```python
#area of rectangle
length = 10
breadth = 20

print(length * breadth)
```

```python
#perimeter of rectangle
length = 10
breadth = 20

print(2*(length * breadth))
```
