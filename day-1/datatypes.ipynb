```python
#different datatypes
name = 'lava'
age = 20
marks = 95
temperature = -5
print(type(name))
print(type(age))
print(type(marks))
print(type(temperature)) 
```

```python
#float 
height = 5.8
price = 99.99
print(type(height))
print(type(price))
```

```python
#complex
x = 3 + 4j
print(x)
print(type(x))
```

```python
#boolean
is_student = True
is_married = False
print(type(is_student))
print(type(is_married))
```

```python
#string
name = "Lava"
city = 'Hyderabad'
print(type(name))
print(type(city))
```

```python
#list 
marks = [90, 85, 95, 88]
print(marks)
print(type(marks))
```

```python
#tuple
colors = ("Red", "Green", "Blue")
print(colors)
print(type(colors))
```

```python
#set
numbers = {10, 20, 30, 20, 10}
print(numbers)
print(type(numbers))
```

```python
#dictionary
student = {
    "name": "Lava",
    "age": 20,
    "marks": 90
}
print(student)
print(type(student))
```

```python
#all in one
age = 20
height = 5.8
number = 2 + 3j
student = True
name = "Lava"
marks = [90, 80, 70]
colors = ("Red", "Green")
unique = {10, 20, 30}
info = {"city": "Hyderabad"}
print(type(age))
print(type(height))
print(type(number))
print(type(student))
print(type(name))
print(type(marks))
print(type(colors))
print(type(unique))
print(type(info))
```
